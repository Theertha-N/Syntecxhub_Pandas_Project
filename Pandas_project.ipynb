STEP 1: IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

STEP 2: LOAD DATASET

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

STEP 7 : EXPLORATORY DATA ANALYSIS

```python
sns.histplot(df['total_bill'])
plt.title("Distribution of Total Bill")
plt.show()
```

- Most bills fall between 10 and 30.

```python
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Total Bill vs Tip")
plt.show()
```

- There is a positive relationship between bill and tip.

```python
sns.barplot(x='day', y='total_bill', data=df)
plt.title("Average Bill by Day")
plt.show()
```

- Weekend days show higher spending.

```python
sns.boxplot(x='sex', y='total_bill', data=df)
plt.title("Spending by Gender")
plt.show()
```

- Spending distribution is similar across genders.

STEP 8 :DATA FILTERING

- Filtering records where total bill is greater than 20

```python
filtered_df = df[df['total_bill'] > 20]
filtered_df.head()
```

STEP 9: SAVE PROCESSED DATA

```python
filtered_df.to_csv('filtered_data.csv', index=False)
```

STEP 10 : FINAL INSIGHTS

## Final Insights

- Most customers spend moderate amounts.
- Higher bills lead to higher tips.
- Weekend shows higher spending trends.
- Data analysis helps understand customer behavior.
